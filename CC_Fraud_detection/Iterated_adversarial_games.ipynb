## Defense

```python
# My functions

def pre_models(x, y):
    forest = RandomForestClassifier(random_state=0, max_depth=5)
    regression = LogisticRegression(random_state=5, max_iter = 500)
    dt = DecisionTreeClassifier(random_state=5, max_depth=5)
    
    forest.fit(x, y)
    regression.fit(x, y)
    dt.fit(x, y)
    
    return [forest, regression, dt]

def retrain(clf, x, y, prev_acc):
    # picking from a bag of models...
    x_tr, x_te, y_tr, y_te = train_test_split(x, y, random_state=42)
    models = pre_models(x_tr, y_tr)
    
    perfs = [balanced_accuracy_score(y_te, model.predict(x_te)) for model in models]
    
    if prev_acc > max(perfs):
        return clf
    else:
        return models[np.argmax(perfs)]
    
```

## Offense

```python
# MY function
def get_best_strategy_indicies(y_actual, y_pred, threshold, y_prob):  # Need to remove the y_prob since it implies that the adversary can access the classification model
    y_actual = np.array(y_actual)
    wrong_indicies = ~np.equal(y_pred,y_actual)
    fn = wrong_indicies & (y_actual & np.logical_not(y_pred))
    fn_index = np.where(fn)[0]
    
    positive, negative = confusion_matrix(y_actual, y_pred)
    assert negative[0] == len(fn_index), "Number of false negative do not match number of false negative indicies"
#     print(type(fn_index))
#     best_strategy_index = []
    
#     # might remove this section
#     for i in fn_index:
#         prob_non_fraud = y_prob[i][0]
#         prob_fraud = y_prob[i][1]
#         if prob_non_fraud > threshold:
#             best_strategy_index.append(i)

#     return np.array(best_strategy_index)
    return np.array(fn_index)

def append_and_smote(best_strat_indicies, x, y, smote):
    y = np.array(y)
    non_fraud_indicies = np.where(np.logical_not(y))[0]
    new_combined_indicies = np.concatenate((non_fraud_indicies, best_strat_indicies), axis=None)

    new_combined_indicies = new_combined_indicies.tolist()
    # To fix float error
    new_combined_indicies = list(map(int, new_combined_indicies))
    
    new_x = x.iloc[new_combined_indicies]
    print("new_combined_indicies: ", new_combined_indicies[0:10])
    new_y = y[new_combined_indicies]
#     new_y = y[list(map(int, new_combined_indicies))]
    
    print("new_y: ", new_y)
    
    x_resampled, y_resampled = smote.fit_resample(new_x, new_y)
    
#     print("y_resampled: ", y_resampled[0:5])
#     print("x_resampled: ", x_resampled[0:5])
    return x_resampled, y_resampled
    
```

## Game

```python
init_df = df[df['Round'] == 0]
x, y = init_df[init_df.columns[0:-2]], init_df[init_df.columns[-2::]]
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(x, y['FRD_IND'], random_state=42)
```

```python
print(Counter(y_tr))
print(Counter(y_te))
```

```python
clf = RandomForestClassifier(random_state=0, max_depth=5)
clf.fit(x_tr, y_tr)
prev_acc = balanced_accuracy_score(y_te, clf.predict(x_te))
```

```python
# intialize stores
balanced_accuracies = [prev_acc]
prev_acc
```

```python
smote = SMOTE(random_state=44)
# it fails at round 9 because the classifier correctly classified all the data, so get_best_strategy_indicies returns empty
# run next 9 rounds
for i in range(9):
    i += 1
    
    # print
    print('Round ' + str(i) + ' Starting ...')
    
    # run offensive strategy (get sample of attacks)
    indicies = get_best_strategy_indicies(y_te, clf.predict(x_te), 0.6, clf.predict_proba(x_te))
    print("get_best_strategy_indicies: ", indicies )
    x_te, y_te = append_and_smote(indicies, x_te, y_te, smote)
    
    print(Counter(y_te))
#     print('... New Attacks Selected by Adversary')
    
    # run defensive strategy (get new classifier)
    clf = retrain(clf, x_te, y_te, prev_acc)
#     print('... New Classifier Selected by Defender')
    
    prev_acc = balanced_accuracy_score(y_te, clf.predict(x_te))
    balanced_accuracies.append(prev_acc)
    
```
